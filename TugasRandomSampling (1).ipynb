```python
import pandas as pd
path = '/content/drive/MyDrive/Statistika/MovieData-2018-2020.csv'
data = pd.read_csv(path)
```

```python
data["genre"].value_counts()
```

```python
filtered_data = data.groupby('genre').filter(lambda x: len(x) > 1)
filtered_data["genre"].value_counts()
```

```python
from sklearn.model_selection import train_test_split
_, stratified_sample = train_test_split(
    filtered_data,
    test_size=0.2,               # Ambil 20% data dari setiap kelas
    stratify=filtered_data['genre'],      # Stratifikasi berdasarkan kolom 'genre'
    random_state=38
)
# Hasil stratified_sample berisi 20% sampel dari setiap genre
stratified_sample
```

```python
stratified_sample.describe()
```

```python
import pandas as pd
import scipy.stats as stats
import math

x_bar = 5.425761e+07
s = 5.927952e+07
n = 30
confidence_level = 0.95

alpha = (1 - confidence_level)
df = n - 1
standard_error = s / math.sqrt(n)
critical_t = stats.t.ppf(1 - alpha/2, df)
margin_of_error = critical_t * standard_error

lower_bound = x_bar - margin_of_error
upper_bound = x_bar + margin_of_error

print("=== RINGKASAN DATA ===")
print("Jumlah Sampel Film (n)   : {}".format(n))
print("Rata-rata Budget (x_bar) : ${:,.2f}".format(x_bar))
print("Standar Deviasi (s)      : ${:,.2f}".format(s))
print("Critical t-value         : {:.3f}".format(critical_t))
print("Margin of Error          : ${:,.2f}".format(margin_of_error))
print("-" * 50)
print("{:.1%} Interval kepercayaan untuk rata-rata populasi adalah:".format(confidence_level))
print("diantara ${:,.2f} and ${:,.2f}".format(lower_bound, upper_bound))
```
